Above is a wikipedia page of hindi movies. We are interested in getting the data of all hindi movies from the FIlms table.

#### Importing the libraries:

```python
from bs4 import BeautifulSoup
import requests #python library to fetch the data in raw format
import re #RegEx can be used to validate all types of character combinations, including special characters like line breaks.
```

#### Fetching data:

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_Hindi_films_of_1980')
response.status_code
```

```python
response.text[:1000]
```

```python
html_text = response.text
data = BeautifulSoup(html_text, 'html.parser')
```

```python
print(data.prettify())
```

```python
headings_before_table = data.find_all('h2')
headings_before_table
```

#### Table data:

```python
table_data = data.find_all('table', class_='wikitable')
len(table_data)
```

```python
table_data
```

```python
table_data = data.find('table',class_='wikitable')
table_data
```

```python
#Extracting rows from the table
rows = table_data.find_all('tr')
rows
```

```python
#storing all the values in a list. Replacing empty values by NA
values = []
for row in rows:
    current_row = []
    for cell in row.find_all('td'):
#         if cell.text != "\n" and cell.text != None:
#             current_row.append(cell.text)
#         else:
#             current_row.append("NA")
          if cell.text == None or cell.text =='':
                current_row.append("NA")
          else:
            current_row.append(cell.text)
        
    values.append(current_row)
values
    
```

#### Converting the required data into a dataframe:

```python
import pandas as pd

df = pd.DataFrame()

for row in values:
    if len(row) != 0:
        df = df.append({'Movie Name' : row[0],
                       'Director' : row[1],
                       'Cast' : row[2],
                       'Genre': row[3]},ignore_index= True)
df.head()
```
